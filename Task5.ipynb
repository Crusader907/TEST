Часть 1: Работа с Polars
Шаг 1.1: Считывание датасета из файла

```python
import polars as pl

# Считываем данные из файла
df = pl.read_csv('train.csv')

# Вывод первых строк
display(df.head())
```

Шаг 1.2: Основная информация о датасете

```python
# Типы данных
print("===Типы данных===")
print(df.dtypes)

# Количество пропусков
print("\n===Количество пропусков===")
display(df.null_count())

# Описательная статистика
print("\n===Описательная статистика===")
display(df.describe())
```

Шаг 1.3: Количество пассажиров каждого класса (Pclass)

```python
pclass_counts = df.get_column('Pclass').value_counts()
print("\n===Количество пассажиров каждого класса===")
display(pclass_counts)
```

Шаг 1.4: Количество выживших мужчин и женщин

```python
survived_by_sex = df.group_by('Sex').agg(pl.col('Survived').sum())
print("\n===Количество выживших мужчин и женщин===")
display(survived_by_sex)
```

Шаг 1.5: Пассажиры старше 44 лет

```python
older_than_44 = df.filter(pl.col('Age') > 44)
print("\n===Пассажиры старше 44 лет===")
display(older_than_44)
```

Часть 2: Ускорение работы с Pandas
Шаг 2.1: Считывание датасета из файла

```python
import pandas as pd

# Считываем данные из CSV-файла
df_pandas = pd.read_csv('train.csv')

# Выводим первые строки датасета для проверки
display(df_pandas.head())
```

Шаг 2.2: Средний возраст и стандартное отклонение с помощью Bottleneck

```python
import bottleneck as bn

mean_age = bn.nanmean(df_pandas['Age'])
std_age = bn.nanstd(df_pandas['Age'])

print(f"\nСредний возраст: {mean_age:.2f}")
print(f"Стандартное отклонение возраста: {std_age:.2f}")
```

Шаг 2.3: Создание нового столбца

```python
# Создаем новый столбец Fare_new
df_pandas['Fare_new'] = df_pandas['Fare'].apply(lambda x: x * 1.3)

# Выводим первые строки для проверки
display(df_pandas[['Fare', 'Fare_new']].head())
```

Часть 3: Оптимизация типов данных в Pandas
Шаг 3.1: Считывание датасета из файла

```python
# Считываем данные из CSV-файла
housing_df = pd.read_csv('Housing.csv')

# Выводим первые строки датасета для проверки
display(housing_df.head())
```

Шаг 3.2: Определение оптимальных типов данных

```python
# Текущее потребление памяти
display(f"Текущее потребление памяти: {housing_df.memory_usage(deep=True).sum() / 1024:.2f} KB")

# Информация о типах данных
display(housing_df.info())

# Описательная статистика для числовых столбцов
display(housing_df.describe())
```

Приведенные выше типы данных можно заменить на более компактные типы:
int8, int16, int32 — для целых чисел.
float16, float32 — для чисел с плавающей точкой.

Для столбца "price":
Тип данных: int64.
Диапазон значений: от минимума до максимума - от 1750000 до 13300000).
Оптимальный тип: int32.
Аналогично для столбца "area" - int 16
"bedrooms" - int 8


Шаг 3.3: Изменение типов данных

```python
# Создаем копию датафрейма
housing_df_optimized = housing_df.copy()

# Преобразуем числовые столбцы
housing_df_optimized['price'] = housing_df_optimized['price'].astype('int32')
housing_df_optimized['area'] = housing_df_optimized['area'].astype('int16')
housing_df_optimized['bedrooms'] = housing_df_optimized['bedrooms'].astype('int8')
housing_df_optimized['bathrooms'] = housing_df_optimized['bathrooms'].astype('int8')
housing_df_optimized['stories'] = housing_df_optimized['stories'].astype('int8')
housing_df_optimized['parking'] = housing_df_optimized['parking'].astype('int8')

# Преобразуем категориальные столбцы
categorical_columns = [
    'mainroad', 'guestroom', 'basement', 'hotwaterheating',
    'airconditioning', 'prefarea', 'furnishingstatus'
]
for col in categorical_columns:
    housing_df_optimized[col] = housing_df_optimized[col].astype('category')

# Проверяем потребление памяти после оптимизации
display(f"Потребление памяти до оптимизации: {housing_df.memory_usage(deep=True).sum() / 1024:.2f} KB")
display(f"Потребление памяти после оптимизации: {housing_df_optimized.memory_usage(deep=True).sum() / 1024:.2f} KB")
```
